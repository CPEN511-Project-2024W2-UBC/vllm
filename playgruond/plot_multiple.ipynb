```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
places_to_chack = {
    'FB_125m' : 'linear_test/FB_125m_2-7.5_0.2_neards/',
    'FB_1.3b' : 'linear_test/FB_1.3b_2-7.5_0.2_neards/',
    'FB_2.7b' : 'linear_test/FB_2.7b_2-7.5_0.2_neards/',
    'DS_1.5b' : 'linear_test/DS_1.5b_2-7.5_0.2_neards/',
}
# playgruond/linear_test
```

```python
places_to_chack = {
    'FB_125m' : 'FB_125m_0-12_0.2_neards/',
    'FB_1.3b' : 'FB_1.3b_0-12_0.2_neards/',
    'FB_2.7b' : 'FB_2.7b_0-12_0.2_neards/',
    'DS_1.5b' : 'DS_1.5b_0-12_0.2_neards/',
}
```

```python
# Initialize a figure for the combined plot
plt.figure(figsize=(10, 6))
columns_to_plot = ['blk_in_count', 'swap_in_count', 'in_spd', 'out_spd']

# Iterate over each column to plot separately
for column in columns_to_plot:
    # Initialize a new figure for each column
    plt.figure(figsize=(10, 6))
    
    # Iterate over each label and directory in places_to_chack
    for label, directory in places_to_chack.items():
        # Read the stats.log file
        file_path = f"{directory}stats.log"
        df = pd.read_csv(file_path)
        
        # Group by 'opt_facotr' and calculate the mean of the other columns
        df_avg = df.groupby('opt_facotr', as_index=False).mean()
        
        # Plot the specified column against 'opt_facotr'
        plt.plot(df_avg['opt_facotr'], df_avg[column], marker='o', label=f'{label}')
    
    # Add title, labels, legend, and grid to the plot
    plt.title(f'{column} vs opt_facotr')
    plt.xlabel('opt_facotr')
    plt.ylabel(column)
    plt.legend()
    plt.grid(True)
    
    # Save the plot for the current column
    plt.savefig(f'{column}_vs_opt_facotr.png')
    
    # Show the plot
    plt.show()
```
